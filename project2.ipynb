```python
import pandas as pd
df_fights = pd.read_csv("ufc-master.csv")
df_fights
```

```python
df_blue_winner = df_fights[df_fights["Winner"] == "Blue"]
df_red_winner = df_fights[df_fights["Winner"] == "Red"]
frames = [df_blue_winner, df_red_winner]

```

```python
blue_heavyweights = df_blue_winner[df_blue_winner["weight_class"] == "Heavyweight"]
red_heavyweights = df_red_winner[df_red_winner["weight_class"] == "Heavyweight"]


```

```python
blue_lightheavyweights = df_blue_winner[df_blue_winner["weight_class"] == "Light Heavyweight"]
red_lightheavyweights = df_red_winner[df_red_winner["weight_class"] == "Light Heavyweight"]


```

```python
blue_middleweights = df_blue_winner[df_blue_winner["weight_class"] == "Middleweight"]
red_middleweights = df_red_winner[df_red_winner["weight_class"] == "Middleweight"]

```

```python
blue_welterweights = df_blue_winner[df_blue_winner["weight_class"] == "Welterweight"]
red_welterweights = df_red_winner[df_red_winner["weight_class"] == "Welterweight"]

```

```python
blue_flyweights = df_blue_winner[df_blue_winner["weight_class"] == "Flyweight"]
red_flyweights = df_red_winner[df_red_winner["weight_class"] == "Flyweight"]

```

```python
average_heavyweight_age = (blue_heavyweights["B_age"].mean() + red_heavyweights["R_age"].mean()) / 2
average_lightheavyweight_age = (blue_lightheavyweights["B_age"].mean() + red_lightheavyweights["R_age"].mean()) / 2
average_middleweight_age =  (blue_middleweights["B_age"].mean() + red_middleweights["R_age"].mean()) / 2
average_welterweight_age = (blue_welterweights["B_age"].mean() + red_welterweights["R_age"].mean()) / 2
average_lightweight_age = (blue_lightweights["B_age"].mean() + red_lightweights["R_age"].mean()) / 2
average_featherweight_age = (blue_featherweights["B_age"].mean() + red_featherweights["R_age"].mean()) / 2
average_bantamweight_age = (blue_bantamweights["B_age"].mean() + red_bantamweights["R_age"].mean()) / 2
average_flyweight_age = (blue_flyweights["B_age"].mean() + red_flyweights["R_age"].mean()) / 2

print("The average heavyweight age is " + str(average_heavyweight_age) + ".")
print("The average light heavyweight age is " + str(average_lightheavyweight_age) + ".")
print("The average middleweight age is " + str(average_middleweight_age) + ".")
print("The average welterweight age is " + str(average_welterweight_age) + ".")
print("The average lightweight age is " + str(average_lightweight_age) + ".")
print("The average featherweight age is " + str(average_featherweight_age) + ".")
print("The average bantamweight age is " + str(average_bantamweight_age) + ".")
print("The average flyweight age is " + str(average_flyweight_age) + ".")



 

```
